````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn
````

```numpy.log1p(arr, out = None, *, where = True, casting = ‘same_kind’, order = ‘K’, dtype = None, ufunc ‘log1p’) ```:

This mathematical function helps user to calculate natural logarithmic value of x+1 where x belongs to all the input array elements.

````python
def softplus(x) :
    return np.log1p(np.exp(x))
````

````python
def forward(x, W1, W2) :
    hidden = np.tanh(x.dot(W1))
    output = hidden.dot(W2)
    mean = output[:2] #since first 2 are mean 
    stddev = softplus(output[2:])
    return mean, stddev
````

````python
#since nn is of size (4, 3, 2)

W1 = np.random.randn(4, 3)
W2 = np.random.rand(3, 2 * 2)

#we need 2 components for the mean and 2 for std dev.
````

````python
#make a random input
x = np.random.randn(4)
print(f'x  = {x}')

#getting the parameters of the gaussian 
mean, stddev = forward(x, W1, W2)
print(f'mean = {mean}')
print(f'stddev  = {stddev}')

#draw samples
samples = mvn.rvs(mean = mean, cov = stddev**2, size = 10000) #used to sample from gaussian pdf

#plot the samples
plt.scatter(samples[:,0], samples[:,1], alpha = 0.5)
plt.show()
````
