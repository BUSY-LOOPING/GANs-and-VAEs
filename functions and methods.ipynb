```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

x = np.linspace(0, 5, 10, endpoint=False)
x
```

```python
y = multivariate_normal.pdf(x, mean=2.5, cov=0.1)
print(y)

fig1 = plt.figure(figsize=(5, 5))
ax = fig1.add_subplot(111)
ax.plot(x, y)
plt.show()
```

```python
y1 = multivariate_normal.rvs(mean=0, cov=2.5, size = 100)
print(y1.std())
```
