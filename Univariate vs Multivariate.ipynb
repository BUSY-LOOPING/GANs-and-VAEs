## Univariate

- Univariate distributions are distributions that describe the probability of a single random variable. 
- For example, the normal distribution is a univariate distribution that describes the probability of a continuous random variable. The normal distribution is often used to model data that is normally distributed, such as height, weight, and blood pressure.
- A linear combination of 2 or more normal random variables is also normally distributed.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a random sample from a normal distribution.
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the data.
plt.hist(data, bins=100)
plt.title('Univariate Normal Distribution')
plt.show()
```

## Multivariate 
- Multivariate distributions are distributions that describe the probability of multiple random variables. 
- For example, the bivariate normal distribution is a multivariate distribution that describes the probability of two continuous random variables. The bivariate normal distribution is often used to model data that is correlated, such as the height and weight of people.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a random sample from a bivariate normal distribution.
data = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0.5], [0.5, 1]], size=1000)

# Plot the data.
plt.scatter(data[:, 0], data[:, 1], s=10)
plt.show()

plt.figure()
plt.hist(data[:, 0], label = 'First Random Variable', bins = 100)
plt.hist(data[:, 1], label = 'Second Random Variable', bins = 100)
plt.legend()
plt.show()
```

```python
print(data.shape)
np.around(np.cov(data.T), decimals=1)
```

## MultiModal vs Multivariate

**Are they the same?**

**No**, multivariate and multimodal distributions are not the same.

A multivariate distribution is a probability distribution that describes the joint behavior of two or more random variables. In other words, it tells us the probability of observing a particular combination of values for the random variables. For example, a multivariate distribution could be used to describe the heights and weights of a population of people.

A multimodal distribution is a probability distribution that has more than one mode. A mode is a point in the distribution where the probability density function (pdf) is highest. For example, a bimodal distribution would have two modes, meaning that there are two values for the random variable that are more likely to occur than any other values.

So, a multivariate distribution can be multimodal, but a multimodal distribution is not necessarily multivariate. For example, a unimodal distribution (a distribution with one mode) can be multivariate, such as the distribution of heights and weights of a population of people.

Here are some examples of multivariate and multimodal distributions:

- Multivariate distribution: The distribution of heights and weights of a population of people.
- Multimodal distribution: The distribution of the prices of houses in a city, which may have multiple peaks, one for each neighborhood.
- Unimodal distribution: The distribution of the ages of students in a class, which will likely have a single peak, around the average age of the students.
