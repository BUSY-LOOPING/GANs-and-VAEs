```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_probability as tfp

N = 10000
mean = np.ones(N) * 5
scale = np.ones(N) * 3

X = tfp.distributions.Normal(loc=mean, scale=scale)
x = X.sample()

print(f'Sample mean : {x.numpy().mean()}')
print(f'Sample stddev : {x.numpy().std()}')

plt.hist(x.numpy(), bins = 20);
```
